## Agrupamento

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep =';')
```

```python
dados.head(10)
```

```python
dados['Bairro'].unique()
```

```python
bairros = list(dados['Bairro'].unique())
```

```python
grupo_bairros = dados.groupby('Bairro')
grupo_bairros.groups
```

```python
grupo_bairros[['Valor']].mean().round(2).head(10)
```

```python
for bairro, data in grupo_bairros:
    print(data[['Bairro', 'Valor']])
```

## Estatísticas descritivas

```python
grupo_bairros['Valor'].describe().round(2).head(10)
```

```python
grupo_bairros['Valor'].aggregate(['min', 'max']).rename(columns = {'min': 'Mínimo', 'max': 'Máximo'}).head(10)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20 , 10))
```

```python
fig = grupo_bairros['Valor'].mean().plot.bar('color = blue')
fig.set_ylabel('Valor do Aluguel R$')
fig.set_title('Valor Médio do Aluguel por Bairro', {'fontsize': 22})
```
