## Tratamento de dados

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep= ';')
dados.head(10)
```

```python
dados.isnull()
```

```python
dados.notnull()
```

```python
dados.info()
```

```python
dados.dropna(subset = ['Valor'], inplace = True)
dados
```

```python
# Neste caso, considera-se que todo imovel do tipo apartamento deva ser cobrado o valor do condomínio
# Assim, todos os apartamentos sem valor de condominio serão apagados da base de dados

selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
dados = dados[~selecao]
dados
```

```python
# Agora atribui-se valor zero para Condomínio e IPTU faltantes

dados = dados.fillna({'Condominio' : 0, 'IPTU' : 0})
dados.head(10)
```

```python
dados.info()
```

```python
dados.to_csv('dados/aluguel_residencial.csv', sep = ';', index = False)
```
